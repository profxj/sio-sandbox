# Line 90

# Load

```python
line = '90'
items = cugn_io.load_line(line)
ds = items['ds']
```

# Find em

```python
xkey = 'SA'
ykey = 'CT'
variable = 'doxy'
#
gd = np.isfinite(ds[xkey]) & np.isfinite(ds[ykey]) & np.isfinite(ds[variable])
```

```python
bad = (ds[xkey] > 34.) & (ds[ykey] > 16.)
```

# Examine

```python
in_bool = gd & bad
all_iz, all_ip = np.where(gd & bad)
```
